```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.date
df
```

```python
df = df.dropna()
```

```python
df['Day'] = df['Day'].apply(lambda x: int(x))
df
```

```python
df
```

```python
df[' Unit_Price '] = df[' Unit_Price '].apply(lambda x: float(x))
df[' Unit_Cost '] = df[' Unit_Cost '].apply(lambda x: float(x))
df[' Profit '] = df[' Profit '].apply(lambda x: float(x))
df[' Cost '] = df[' Cost '].apply(lambda x: float(x))
df['Revenue'] = df['Revenue'].apply(lambda x: float(x))
df
```

```python
df[['Customer_Age','Order_Quantity', ' Unit_Cost ',' Unit_Price ', ' Profit ', ' Cost ', 'Revenue']].corr()
```

```python
df['Age_Group'].value_counts()
```

```python
df['Age_Group'] = ''  # Initialize the 'Age_Group' column

for i, x in df['Customer_Age'].items():
    if x < 25:
        df.loc[i, 'Age_Group'] = 'Youth'
    elif x >= 25 and x <= 34:
        df.loc[i, 'Age_Group'] = 'Young Adults'
    elif x >= 35 and x <= 64:
        df.loc[i, 'Age_Group'] = 'Adults'
    else:
        df.loc[i, 'Age_Group'] = 'Seniors'
```

```python
df['Product_Description'].value_counts()
```

```python
df[['Product_Description','Size']] = df['Product_Description'].str.split(' ,', n=2, expand=True)
```

```python
df
```

```python
df[['Product_Description','Color']] = df['Product_Description'].str.split(' ', n=1, expand=True)
```
