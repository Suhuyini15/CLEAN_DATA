```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./ethereum.csv')
df
```

```python
df.head(30)
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['DateTime (UTC)'] = pd.to_datetime(df['DateTime (UTC)'])
```

```python
df['Date'] = df['DateTime (UTC)'].dt.date
df['Time(UTC)'] = df['DateTime (UTC)'].dt.time
df
```

```python
df['UnixTimestamp'] = pd.to_datetime(df['UnixTimestamp'],unit='ms')
df
```

```python
df = df.drop(columns='DateTime (UTC)')
df
```

```python
df
```

```python
df['VALUE_IN(USD)'] = (df['Value_IN(ETH)']*df['Historical $Price/Eth'])
df
```

```python
df['VALUE_OUT(USD)'] = (df['Value_OUT(ETH)']*df['Historical $Price/Eth'])
df
```

```python
stats_summary = df[['Value_IN(ETH)', 'Value_OUT(ETH)', 'TxnFee(ETH)', 'TxnFee(USD)', 'Historical $Price/Eth']].describe()
```

```python
stats_summary
```

```python
correlation = df[['Value_IN(ETH)', 'Value_OUT(ETH)', 'TxnFee(ETH)', 'TxnFee(USD)', 'Historical $Price/Eth']].corr()
correlation
```
